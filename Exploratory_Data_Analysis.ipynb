```python
# Set a name for a new dataset after applying filters
#data = df[(df['Event distance/length'].isin(['50mi','50km'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
data = df[(df['Event distance/length'].isin(['50mi','50km'])) & (df['Year of event'] == 2020) & (df['Event name'].str.contains('USA'))]
```

```python
data.shape
```

```python
# Remove 'USA' in Event name
# data['Event name'].apply(lambda x: x.split('(')[0])
data['Event name'].str.split('(').str.get(0)
```

```python
data['Event name'] = data['Event name'].str.split('(').str.get(0)
```

```python
data.head()
```

```python
# Create a new column based on Athlete age
data['Athlete age'] = 2020 - df['Athlete year of birth']
```

```python
data.head()
```

```python
# Remove h in Athlete Performance
#data['Athlete performance'].apply(lambda y: y.split(' ')[0])
data['Athlete performance'].str.split(' ').str.get(0)
```

```python
data['Athlete performance'] = data['Athlete performance'].str.split(' ').str.get(0)
```

```python
data.head()
```

```python
# Drop columns: Club, Country, Year of birth, Age category
data = data.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'], axis = 1)
```

```python
# @title Event distance/length
data.groupby('Event distance/length').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```
